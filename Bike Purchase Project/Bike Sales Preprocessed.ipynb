```python
import pandas as pd
import numpy as np
```

```python
raw_csv = pd.read_csv("Bike Sales Raw.csv")
```

```python
df = raw_csv
df
```

```python
df['Purchased Bike'].value_counts()
```

```python
df.info()
```

```python
df = df.drop(['ID'], axis=1)
```

```python
df.columns.values
```

```python
df['Marital Status'] = df['Marital Status'].map({'M':1, 'S':0})
```

```python
df['Gender'] = df['Gender'].map({'M':1, 'F':0})
```

```python
df['Education'] = df['Education'].map({'Partial High School':0,'High School':1, 'Partial College':2, 
                                                 'Bachelors':3, 'Graduate Degree':4})
```

```python
df['Home Owner'] = df['Home Owner'].map({'Yes':1, 'No':0})
```

```python
df['Commute Distance'] = df['Commute Distance'].map({'0-1 Miles':0, '1-2 Miles':1, '2-5 Miles':2, 
                                                               '5-10 Miles':3, '10+ Miles':4})
```

```python
df['Region'] = df['Region'].map({'Europe':0, 'Pacific':1, 'North America':2})
```

```python
df['Purchased Bike'] = df['Purchased Bike'].map({'Yes':1, 'No':0})
```

```python
df
```

```python
Occupation_column = pd.get_dummies(raw_csv["Occupation"])
Occupation_column
```

```python
Occupation_column.columns.values
```

```python
df = df.drop(['Occupation'], axis=1)
```

```python
df
```

```python
df['Purchased Bike'].value_counts()
```

```python
df = pd.concat([df, Occupation_column], axis = 1)
df.columns.values
```

```python
column_names = ['Marital Status', 'Gender', 'Income', 'Children', 'Education', 'Clerical', 'Management', 
                'Manual', 'Professional', 'Skilled Manual','Home Owner', 'Cars', 'Commute Distance', 
                'Region', 'Age', 'Purchased Bike']
```

```python
df = df[column_names]
```

```python
df
```

```python
df['Purchased Bike'].value_counts()
```

```python
df_preprocessed = df
```

```python
df_preprocessed.to_csv('Bike_Sales_preprocessed.csv')
```
